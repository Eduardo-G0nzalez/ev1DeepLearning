```
# Esta variable contiene un mapeo de nГәmero de clase a personaje.
# Utilizamos sГіlo los 18 personajes del dataset que tienen mГЎs imГЎgenes.
MAP_CHARACTERS = {
    0: 'abraham_grampa_simpson', 1: 'apu_nahasapeemapetilon', 2: 'bart_simpson',
    3: 'charles_montgomery_burns', 4: 'chief_wiggum', 5: 'comic_book_guy', 6: 'edna_krabappel',
    7: 'homer_simpson', 8: 'kent_brockman', 9: 'krusty_the_clown', 10: 'lisa_simpson',
    11: 'marge_simpson', 12: 'milhouse_van_houten', 13: 'moe_szyslak',
    14: 'ned_flanders', 15: 'nelson_muntz', 16: 'principal_skinner', 17: 'sideshow_bob'
}

# Vamos a standarizar todas las imГЎgenes a tamaГұo 64x64
IMG_SIZE = 64
```

```
def load_train_set(dirname, map_characters, verbose=True):
    """Esta funciГіn carga los datos de training en imГЎgenes.

    Como las imГЎgenes tienen tamaГұos distintas, utilizamos la librerГӯa opencv
    para hacer un resize y adaptarlas todas a tamaГұo IMG_SIZE x IMG_SIZE.

    Args:
        dirname: directorio completo del que leer los datos
        map_characters: variable de mapeo entre labels y personajes
        verbose: si es True, muestra informaciГіn de las imГЎgenes cargadas

    Returns:
        X, y: X es un array con todas las imГЎgenes cargadas con tamaГұo
                IMG_SIZE x IMG_SIZE
              y es un array con las labels de correspondientes a cada imagen
    """
    X_train = []
    y_train = []
    for label, character in map_characters.items():
        files = os.listdir(os.path.join(dirname, character))
        images = [file for file in files if file.endswith("jpg")]
        if verbose:
          print("Leyendo {} imГЎgenes encontradas de {}".format(len(images), character))
        for image_name in images:
            image = cv2.imread(os.path.join(dirname, character, image_name))
            X_train.append(cv2.resize(image,(IMG_SIZE, IMG_SIZE)))
            y_train.append(label)
    return np.array(X_train), np.array(y_train)
```

```
def load_test_set(dirname, map_characters, verbose=True):
    """Esta funciГіn funciona de manera equivalente a la funciГіn load_train_set
    pero cargando los datos de test."""
    X_test = []
    y_test = []
    reverse_dict = {v: k for k, v in map_characters.items()}
    for filename in glob.glob(dirname + '/*.*'):
        char_name = "_".join(filename.split('/')[-1].split('_')[:-1])
        if char_name in reverse_dict:
            image = cv2.imread(filename)
            image = cv2.resize(image, (IMG_SIZE, IMG_SIZE))
            X_test.append(image)
            y_test.append(reverse_dict[char_name])
    if verbose:
        print("LeГӯdas {} imГЎgenes de test".format(len(X_test)))
    return np.array(X_test), np.array(y_test)
```

```
# Cargamos los datos. Si no estГЎs trabajando en colab, cambia los paths por
# los de los ficheros donde hayas descargado los datos.
DATASET_TRAIN_PATH_COLAB = "/root/.keras/datasets/simpsons"
DATASET_TEST_PATH_COLAB = "/root/.keras/datasets/simpsons_testset"

X, y = load_train_set(DATASET_TRAIN_PATH_COLAB, MAP_CHARACTERS)
X_t, y_t = load_test_set(DATASET_TEST_PATH_COLAB, MAP_CHARACTERS)
```

## DefiniciГіn de modelos base

```
from tensorflow.keras.layers import Input, Dense, Flatten, Dropout, Activation, Conv2D, MaxPooling2D
from tensorflow.keras.models import Sequential

IMG_SIZE = 64

# --- Modelo A: MLP (Arquitectura Base) ---
model_mlp = Sequential([
    Input(shape=(IMG_SIZE, IMG_SIZE, 3)),
    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.3), # RegularizaciГіn bГЎsica
    Dense(128, activation='relu'),
    Dense(18, activation='softmax') # 18 personajes
])
model_mlp.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# --- Modelo B: CNN (Arquitectura Base) ---
model_cnn = Sequential([
    Input(shape=(IMG_SIZE, IMG_SIZE, 3)),
    Conv2D(32, (3, 3), padding='same', activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, (3, 3), padding='same', activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(18, activation='softmax')
])
model_cnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

En esta fase, construimos dos arquitecturas iniciales para comparar empГӯricamente el rendimiento de un modelo lineal frente a uno especializado en imГЎgenes. Ambos modelos utilizan una capa `Input` con una forma de (`64, 64, 3`), que corresponde a las dimensiones de las imГЎgenes (64x64 pГӯxeles) y los 3 canales de color RGB.

**Modelo A: PerceptrГіn Multicapa (MLP)**

- **PropГіsito:** ActГәa como nuestro modelo base (baseline).
- **Estructura:** La capa `Flatten` "aplasta" la imagen 2D en un vector unidimensional de $64 \times 64 \times 3 = 12.288$ elementos. Luego, este vector pasa por capas Dense (totalmente conectadas).
- **LimitaciГіn:** El MLP pierde toda la informaciГіn espacial (la relaciГіn entre pГӯxeles cercanos), lo que lo hace ineficiente para detectar patrones visuales complejos como formas o texturas.

**Modelo B: Red Neuronal Convolucional (CNN)**

- **PropГіsito:** Demostrar cГіmo la extracciГіn de caracterГӯsticas mejora la clasificaciГіn de imГЎgenes.
- **Estructura:** Utiliza capas `Conv2D` que aplican filtros (de $3 \times 3$) para detectar caracterГӯsticas como bordes o colores, manteniendo la estructura espacial 2D de la imagen. Las capas `MaxPooling2D` reducen la dimensionalidad espacial, haciГ©ndolo computacionalmente eficiente y extrayendo las caracterГӯsticas mГЎs dominantes. Finalmente, se aplasta la salida (`Flatten`) y pasa por capas `Dense` para la clasificaciГіn final.

**Elementos comunes:**

- **RegularizaciГіn:** Ambos usan capas `Dropout` (0.3), que "apagan" aleatoriamente el 30% de las neuronas durante el entrenamiento, obligando a la red a no depender de caracterГӯsticas especГӯficas y evitando el sobreajuste.
- **Clasificador Final:** La Гәltima capa `Dense` tiene 18 neuronas (una por cada personaje) y utiliza la funciГіn de activaciГіn `softmax`, la cual convierte los valores de salida en probabilidades que suman 1.
- **CompilaciГіn:** Ambos se compilan con el optimizador `adam` y la funciГіn de pГ©rdida `categorical_crossentropy`.

## Entrenamiento Comparativo

**AnГЎlisis del Resultado:**

Al comparar los modelos tras solo 10 Г©pocas, comprobamos empГӯricamente la superioridad de la arquitectura convolucional (CNN) frente a las redes densas tradicionales (MLP) para la clasificaciГіn de imГЎgenes.


- **MLP (Aprox. 29% de precisiГіn):** El bajo rendimiento del PerceptrГіn Multicapa se debe a que su primera capa (`Flatten`) "aplasta" la imagen, perdiendo toda la jerarquГӯa y estructura espacial de los pГӯxeles. El modelo no puede entender quГ© pГӯxel estГЎ al lado de cuГЎl, limitГЎndose a memorizar colores estГЎticos.


- **CNN Base (Aprox. 79% de precisiГіn):** Por el contrario, la Red Neuronal Convolucional alcanza rГЎpidamente un alto nivel de precisiГіn. Esto ocurre porque sus filtros (`Conv2D`) extraen activamente caracterГӯsticas espaciales locales (como bordes, curvas y paletas de colores especГӯficas de cada personaje), manteniendo la estructura 2D de la imagen en las primeras capas.

## CNN Optimizada con RegularizaciГіn

```
# --- MODELO C: CNN Optimizada ---
model_cnn_opt = Sequential([
    Input(shape=(IMG_SIZE, IMG_SIZE, 3)),

    # Bloque 1
    Conv2D(32, (3, 3), padding='same'),
    Activation('relu'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.25),

    # Bloque 2
    Conv2D(64, (3, 3), padding='same'),
    Activation('relu'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.30),

    # Bloque 3
    Conv2D(128, (3, 3), padding='same'),
    Activation('relu'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.40),

    # Clasificador Final
    Flatten(),
    Dense(512, activation='relu'),
    BatchNormalization(),
    Dropout(0.50),
    Dense(18, activation='softmax')
])

# Ajustamos el Learning Rate del optimizador
opt = Adam(learning_rate=0.001)
model_cnn_opt.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])

model_cnn_opt.summary()
```

En este bloque definimos nuestro modelo principal, implementando tГ©cnicas avanzadas de Deep Learning para estabilizar el aprendizaje, evitar el overfitting y alcanzar nuestra meta de precisiГіn (superar el 85%).

**ExtracciГіn JerГЎrquica de CaracterГӯsticas (Bloques Conv2D + MaxPooling):**

- El modelo estГЎ estructurado en tres bloques convolucionales que aumentan en profundidad (32 $\rightarrow$ 64 $\rightarrow$ 128 filtros).

- **JustificaciГіn:** Las primeras capas aprenden a detectar patrones simples (bordes, lГӯneas), mientras que las capas mГЎs profundas combinan esos patrones para reconocer formas complejas (los ojos de Homer, el pelo de Marge). Las capas de `MaxPooling2D` reducen secuencialmente el tamaГұo de la imagen ($64 \times 64 \rightarrow 32 \times 32 \rightarrow 16 \times 16 \rightarrow 8 \times 8$), destilando la informaciГіn mГЎs importante y reduciendo la carga computacional.

**Batch Normalization (NormalizaciГіn por Lotes):**

- AГұadimos esta capa despuГ©s de cada activaciГіn `relu` y antes del clasificador final.

- **JustificaciГіn:** Normaliza las salidas de la capa anterior, asegurando que los valores no se disparen ni se desvanezcan. Esto estabiliza enormemente la red, permitiendo que aprenda mucho mГЎs rГЎpido y de manera mГЎs segura.

**Dropout Progresivo (RegularizaciГіn):**

- Implementamos tasas de Dropout que van aumentando conforme la red se hace mГЎs profunda ($0.25 \rightarrow 0.30 \rightarrow 0.40 \rightarrow 0.50$).

- **JustificaciГіn:** A medida que avanzamos en la red, hay mГЎs filtros y, por ende, mayor riesgo de que el modelo "memorice" (sobreajuste) las imГЎgenes de entrenamiento. Al "apagar" aleatoriamente un porcentaje creciente de neuronas en cada capa, obligamos a la red a generalizar basГЎndose en mГәltiples caracterГӯsticas y no en un solo pГӯxel o patrГіn.

**Clasificador Final y ParГЎmetros:**

- La capa `Flatten` convierte los mapas de caracterГӯsticas ($8 \times 8 \times 128$) en un vector unidimensional de 8.192 elementos. Este vector se conecta a una capa densa gigante de 512 neuronas.

- El resumen del modelo (`summary`) revela que la red tiene ~4.3 millones de parГЎmetros entrenables. La inmensa mayorГӯa de estos (mГЎs de 4.1 millones) provienen de esta conexiГіn entre el `Flatten` y la primera capa `Dense`.

**Optimizador Controlado:**

- Utilizamos `Adam`, pero esta vez declaramos explГӯcitamente su tasa de aprendizaje (`learning_rate=0.001`). Esto nos da control total sobre quГ© tan grandes son los "pasos" que da el modelo para ajustar sus pesos.

### Entrenamiento de CNN

```
EPOCHS_OPT = 25
BATCH_SIZE = 64

print("Entrenando CNN Optimizada...")
history_cnn_opt = model_cnn_opt.fit(
    X_train, y_train,
    epochs=EPOCHS_OPT,
    batch_size=BATCH_SIZE,
    validation_data=(X_val, y_val),
    verbose=1
)

# GrГЎfico para evidenciar la mejora con optimizaciГіn (Requisito IP 1.3.2)
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 5))

# GrГЎfico de Accuracy
plt.subplot(1, 2, 1)
plt.plot(history_cnn_opt.history['accuracy'], label='Train Accuracy')
plt.plot(history_cnn_opt.history['val_accuracy'], label='Validation Accuracy', linewidth=2)
plt.title('CNN Optimizada: PrecisiГіn (Accuracy)')
plt.xlabel('Гүpocas')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

# GrГЎfico de Loss
plt.subplot(1, 2, 2)
plt.plot(history_cnn_opt.history['loss'], label='Train Loss')
plt.plot(history_cnn_opt.history['val_loss'], label='Validation Loss', linewidth=2)
plt.title('CNN Optimizada: PГ©rdida (Loss)')
plt.xlabel('Гүpocas')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

print(f"Accuracy Final (CNN Optimizada): {history_cnn_opt.history['val_accuracy'][-1]:.4f}")
```

# Desarrollo con MLP (CorrecciГіn)

## MLP Optimizado

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam

print("Construyendo MLP Optimizado...")

# DefiniciГіn del modelo MLP
model_mlp_final = Sequential([
    # 1. Capa de entrada: Aplanamos la imagen (64x64x3 -> 12.288)
    Flatten(input_shape=(64, 64, 3)),

    # 2. Capas Ocultas Anchas con OptimizaciГіn (Batch Norm + Dropout)
    Dense(1024, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),

    Dense(512, activation='relu'),
    BatchNormalization(),
    Dropout(0.4),

    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),

    # 3. Capa de salida (18 clases)
    Dense(18, activation='softmax')
])

# CompilaciГіn
model_mlp_final.compile(optimizer=Adam(learning_rate=0.001),
                        loss='categorical_crossentropy',
                        metrics=['accuracy'])

model_mlp_final.summary()
```

## Entrenamiento de MLP

Primero probaremos con las mismas 25 Г©pocas que usamos en el entrenamiento del CNN optimizado

```
import numpy as np

print("Entrenando MLP por 25 Г©pocas...")

# 1. Entrenamiento
history_mlp_final = model_mlp_final.fit(
    X_train, y_train,
    epochs=25,
    batch_size=64,
    validation_data=(X_val, y_val),
    verbose=1
)

# 2. Predicciones (Guardamos las variables en memoria para la siguiente celda)
y_pred_probs_mlp = model_mlp_final.predict(X_val)
y_pred_classes_mlp = np.argmax(y_pred_probs_mlp, axis=1)
y_true_classes = np.argmax(y_val, axis=1)

print("\nEntrenamiento finalizado.")
```

**AnГЎlisis del Entrenamiento: Inestabilidad y Sobreajuste Severo:**

Los registros de entrenamiento de estas 25 Г©pocas exponen de manera clara la deficiencia estructural del PerceptrГіn Multicapa (MLP) para este problema. Podemos observar dos fenГіmenos crГӯticos:

1. **Ausencia de Convergencia (Inestabilidad):** A diferencia de la CNN, la precisiГіn de validaciГіn (`val_accuracy`) nunca muestra un crecimiento constante. Rebota errГЎticamente entre el 27% y el 46%, demostrando que el modelo estГЎ "adivinando" sin lograr asentar un aprendizaje real.
2. **Sobreajuste (Overfitting) Evidente:** Mientras la precisiГіn de entrenamiento (`accuracy`) logra trepar engaГұosamente por encima del 72%, el error de validaciГіn (`val_loss`) no disminuye, sino que fluctГәa y termina disparГЎndose en la Гәltima Г©poca (llegando a 3.62).

**ConclusiГіn empГӯrica:** El modelo se rinde al intentar extraer caracterГӯsticas lГіgicas y opta por memorizar los pГӯxeles exactos del set de entrenamiento. Al evaluar imГЎgenes nuevas (validaciГіn), su falta de comprensiГіn geomГ©trica y espacial queda expuesta, confirmando que aumentar el nГәmero de Г©pocas no solucionarГӯa el problema, sino que agravarГӯa la memorizaciГіn.
